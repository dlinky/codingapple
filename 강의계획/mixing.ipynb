```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()
driver.get('https://www.youtube.com/playlist?list=PLfjEYxTqMdg3-4pGI5MV8GJiOPkG5TTqW')
```

```python
class_dict = dict()

running_times = driver.find_elements(By.CLASS_NAME, "badge-shape-wiz__text")
titles = driver.find_elements(By.ID, "video-title")

for idx in range(len(running_times)):
  class_dict[idx] = [running_times[idx].text, titles[idx].text]
  print(running_times[idx].text, titles[idx].text)
```

```python
df = pd.DataFrame(class_dict).T
df
```

```python
df.to_excel('mixing class.xlsx')
```
